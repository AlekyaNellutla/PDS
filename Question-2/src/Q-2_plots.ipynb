```python
df.plot(kind="scatter", x="reading score", y="writing score")
```

```python
sns.FacetGrid(df, hue="gender", size=5) \
   .map(plt.scatter, "reading score", "writing score") \
   .add_legend()
```

```python
student_perf_col = df[['race/ethnicity','lunch','test preparation course','parental level of education']]
```

```python
fig , ax = plt.subplots(2,2,figsize = (15,10))     # set up 2 x 2 frame count plot with figsize 10 x 10
for i , subplots in zip (student_perf_col, ax.flatten()):  
  ax = sns.countplot(student_perf_col[i],hue = df['gender'],ax = subplots, palette = 'BuPu')

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")  #it will rotate text on x axis

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue="gender", size=3)
```
